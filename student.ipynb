Exploration of the Dataset;

```python
rotten["runtime"] = rotten["runtime"].str.replace("minutes", '').astype(float)
```

```python
#calculate the IQR and identify potential outliers in the 'runtime' column
low_quant = rotten["runtime"].quantile(0.25)
upp_quant = rotten["runtime"].quantile(0.75)

IQR = upp_quant - low_quant

low_limit = low_quant - 1.5 * IQR
upp_limit = upp_quant + 1.5 * IQR

rotten["runtime"].clip(lower = low_limit, upper = upp_limit, inplace=True)
```
